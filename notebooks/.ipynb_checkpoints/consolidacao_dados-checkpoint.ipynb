```python
import pandas as pd
from pathlib import Path


pasta_dados = Path("../datas")

arquivos_csv = list(pasta_dados.glob("*.csv"))

print("Arquivos encontrados: ")

for arq in arquivos_csv:
    print(' -',arq.name)


#Leitura e consolidação dos CSVs
lista_df = []
for arquivo in arquivos_csv:

    df_temp = pd.read_csv(arquivo,sep=';',encoding='utf8')

    #Nova coluna de indentificação do snapshot
    #Usando o nome do arquivo como identificador
    df_temp['snapshot'] = arquivo.stem

    #Converter valores das colunas de data para o tipo datetime
    #alalise a lógica que você critou referenciando o for com os nomes das colunas padrão, caso isso seja alterado terão que ser alterados aqui
    for col in ['data_vencimento','data_pagamento','data_liquidacao']:
        if col in df_temp.columns:
            df_temp[col] = pd.to_datetime(df_temp[col],errors='coerce')


    #Criando uma coluna de competência a partir de 'fatura-mes' e 'fatura-ano'

    if 'fatura_mes' in df_temp.columns and 'fatura_ano' in df_temp.columns:
        df_temp['competencia'] = pd.to_datetime(df_temp['fatura_ano'].astype(str) + '-' + df_temp['fatura_mes'].astype(str) + '-01',errors='coerce')


    #Adicionando o df que tratado na lista de df
    lista_df.append(df_temp)


#Consolidando todos os df tratados (Snapshots)

df_consolidado = pd.concat(lista_df,ignore_index=True)


print('Visualizando as primeiras linhas do DataFrame consolidado: ')
display(df_consolidado.tail())


Path("../datas/df_consolidado.csv").write_text(df_consolidado.to_csv(sep=";",encoding='utf8',index=False))

```
